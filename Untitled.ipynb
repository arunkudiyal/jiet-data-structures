```python
import sys
```

```python
class MinHeap:
    
    def __init__(self, maxsize):
        self.maxsize = maxsize
        self.size = 0
        self.Heap = [0] * (self.maxsize + 1)
        self.Heap[0] = (-1 * sys.maxsize)
        self.FRONT = 1
        
    def parent(self, index):
        return index//2
    
    def leftChild(self, index):
        return 2*index
    
    def rightChild(self, index):
        return (2*index) + 1
    
    def isLeaf(self, index):
        if index >= (self.size//2) and index <= self.size:
            return True
        return False
    
    def swap(self, index1, index2):
        self.Heap[index1], self.Heap[index2] = self.Heap[index2], self.Heap[index1]
        
    
    def minHeapify(self, index):
        # If the node is a non-leaf node and is greater than any of its child
        if not self.isLeaf(index):
            if(self.Heap[index] > self.Heap[self.rightChild(index)] or self.Heap[index] > self.Heap[self.leftChild(index)]):
                
                # swap the left child and heapify the left child
                if self.Heap[self.leftChild(index)] < self.Heap[self.rightChild(index)]:
                    self.swap(index, self.leftChild(index))
                    self.minHeapify(self.leftChild(index))
                    
                else:
                    self.swap(index, self.rightChild(index))
                    self.minHeapify(self.rightChild(index))
                    
    def insert(self, data):
        if self.size >= self.maxsize:
            print("HEAP IS FULL!")
            return
        self.size += 1
        self.Heap[self.size] = data
        temp = self.size
        
        while self.Heap[temp] < self.Heap[self.parent(temp)]:
            self.swap(temp, self.parent(temp))
            temp = self.parent(temp)
            
    
    def display(self):
        for i in range(1, (self.size//2)+1):
            print("Parent: ", str(self.Heap[i]), 
                 "Left Child: ", str(self.Heap[2*i]),
                 "Right Child: ", str(self.Heap[2*i+1]))
           
    # build the minHeap
    def minHeap(self):
        for i in range((self.size//2), 0, -1):
            self.minHeapify(i)
            
    def remove(self):
        deleted = self.Heap[self.FRONT]
        self.Heap[self.FRONT] = self.Heap[self.size]
        self.size -= 1
        self.minHeapify(self.FRONT)
        return deleted
            

print('The minHeap is ')
minHeap = MinHeap(15)
minHeap.insert(5)
minHeap.insert(3)
minHeap.insert(17)
minHeap.insert(10)
minHeap.insert(84)
minHeap.insert(19)
minHeap.insert(6)
minHeap.insert(22)
minHeap.insert(9)
minHeap.minHeap()
    
minHeap.display()

print("Deleted the node: ", minHeap.remove(), "\n")

print("After deletion, Heap is as follow:\n")
minHeap.display()
```
