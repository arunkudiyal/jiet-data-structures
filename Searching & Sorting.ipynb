```python
# SEARCHING

# 1. linear serach

def linear_search(data, x):
    # loop accross and access each data
    for i in range(len(data)):
        if(data[i] == x):
            return i
    return -1

# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

x = int(input("Enter the data that you are searching for: "))
index = linear_search(dataSet, x)

if(index == -1):
    print("Data", x, " is not avaialable in the data-set")
else:
    print("Data", x, "is in the index", index)
```

```python
# 2. binary search

# the data set should be sorted

def binary_search(data, start, end, x):
    # Base Condition
    if(start < end):
        mid = (start + end) // 2;
        # check if the mid is the position that i am looking for
        #case 1
        if(data[mid] == x):
            return mid;

        #case 2 
        elif(data[mid] > x):
            return binary_search(data, start, mid-1, x)

        else:
            return binary_search(data, mid+1, end, x)
    else:
        return -1
    

# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

x = int(input("Enter the data that you are searching for: "))
index = binary_search(dataSet, 0, n-1, x)

if(index == -1):
    print("Data", x, " is not avaialable in the data-set")
else:
    print("Data", x, "is in the index", index)
```

```python
# iterative approach for binary search

def iterative_binary_search(data, x):
    start = 0
    n = len(data)
    end = n-1
    mid = 0
    
    while start <= end:
        mid = (start + end) // 2
        if(data[mid] == x):
            return mid
        
        # 
        elif(data[mid] > x):
            end = mid-1
        
        else:
            start = mid+1
    
    return -1

# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

x = int(input("Enter the data that you are searching for: "))
index = iterative_binary_search(dataSet, x)

if(index == -1):
    print("Data", x, " is not avaialable in the data-set")
else:
    print("Data", x, "is in the index", index)
            
```

```python
# SORTING

# 1. Bubble Sorting

def bubble_sort(data):
    n = len(data)
    
    for i in range(n-1):
        for j in range(0, n-i-1):
            if(data[j] > data[j+1]):
                #swap(data[j], data[j+1])
                data[j], data[j+1] = data[j+1], data[j]
                
# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

print("Data before sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()

bubble_sort(dataSet)

print("Data after sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()
```

```python
# 2. Insertion Sort

def insertion_sort(data):
    n = len(data)
    
    for i in range(1, n):
        key = data[i]
        j = i-1
        
        while j >= 0 and key < data[j]:
            data[j+1] = data[j]
            j -= 1
        data[j+1] = key

# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

print("Data before sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()

insertion_sort(dataSet)

print("Data after sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()
```

```python
# 3. Merge Sort

def merge(data, start, mid, end):
    nL = mid - start + 1
    nR = end - mid
    
    # create the temp data sets L and R
    L = [0] * (nL)
    R = [0] * (nR)
    
    # copy the data from array to L & R
    for i in range(0, nL):
        L[i] = data[start + i]
        
    for j in range(0, nR):
        R[j] = data[mid + 1 + j]
        
    
    # Merge L and R | i is for L | j is for R | k is for A
    i = 0
    j = 0
    k = start
    
    while i < nL and j < nR:
        if(L[i] <= R[j]):
            data[k] = L[i]
            i += 1
            k += 1
        else:
            data[k] = R[j]
            j += 1
            k += 1
    
    # copy the remaining data from the L or R
    while i < nL:
        data[k] = L[i]
        i += 1
        k += 1
    
    while j < nR:
        data[k] = R[j]
        j += 1
        k += 1
   

def merge_sort(data, start, end):
    if start < end:
        # mid value
        mid = (start + (end-1)) // 2
        
        # break the problem
        merge_sort(data, start, mid)
        merge_sort(data, mid+1, end)
        merge(data, start, mid, end)

# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

print("Data before sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()

merge_sort(dataSet, 0, n-1)

print("Data after sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()        
```

```python
# Quick Sort

def partition(data, start, end):
    pivot = data[end]
    pIndex = start
    
    for i in range(start, end):
        if(data[i] <= pivot):
            #swap(data[i], data[pIndex])
            data[i], data[pIndex] = data[pIndex], data[i]
            pIndex = pIndex + 1
            
    data[pIndex], data[end] = data[end], data[pIndex] 
    return pIndex


def quick_sort(data, start, end):
    # corner_case
    if(len(data) == 1):
        return data
    if(start < end):
        pi = partition(data, start, end)
        quick_sort(data, start, pi-1)
        quick_sort(data, pi+1, end)
        
# driver code
dataSet = []
n = int(input("Enter the size of the data set: "))
for i in range(0, n):
    element = int(input())
    dataSet.append(element)

print("Data before sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()

quick_sort(dataSet, 0, n-1)

print("Data after sorting: ")
for i in range(n):
    print(dataSet[i], end=" ")
print()
```

```python
# STACKS & QUEUES

# Stacks [LIFO]

# 1. Stacks using a List
stack = []

stack.append("Hello")
stack.append("how")
stack.append("are")
stack.append("you")
print(stack)
# append() <-> push()

print(stack.pop())
print(stack.pop())
# pop() <-> pop()

# display()
print(stack)
```

```python
# Queues [FIFO]

# Queues using a List
queue = []

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)
# append <-> enQueue()
print(queue)

print(queue.pop(0))
print(queue.pop(0))
# pop(0) <-> deQueue()

#display
print(queue)
```

```python
# Collections

# deque
from collections import deque
queue = deque([])

queue.append(10)
queue.append(20)
queue.append(30)
queue.append(40)
queue.append(50)
queue.append(60)
queue.append(70)
print(queue)

# you can create a stack or a queue using a deque

# stack using deque
print(queue.pop())
print(queue.pop())
print("The Stack is: ", queue)

# queue using deque
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())
print("The Queue is: ", queue)

```
