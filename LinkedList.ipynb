```python
# container which could tell how the Node is looking like
class Node:
    # constructor
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    # constructor of LL
    def __init__(self):
        self.head = None
        
    # ADD METHODS
    
    # 1. Add at the begining of the LL
    def addBegining(self, data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode
        
    # 2. Add at the end of the LL
    def addEnd(self, data):
        # create the Node, paste the data
        newNode = Node(data)
        
        # check if the LL is empty
        if self.head is None:
            # this Node will be the I node
            self.head = newNode
            return
        
        # LL already exists
        # traverse to the end, and change the next of the last node
        temp = self.head
        #traverse the LL
        while(temp.next is not None):
            temp = temp.next
        temp.next = newNode
        
    # 3. Add at the middle of a given node in the LL
    def addMid(self, midNode, data):
        # condition -> midNode == Node
        if(midNode == None):
            print("Node is NOT available")
            return
        
        # create the node, and do the linking
        newNode = Node(data)
        newNode.next = midNode.next
        midNode.next = newNode
        
    
    # DELETE METHOD
    
    def remove(self, data):
        temp = self.head
        
        # LL does not exist
        if(temp is not None):
            # case 1: remove the first Node
            if(self.head.data == data):
                self.head = temp.next
                temp = None
                return
            
        while(temp is not None):
            # search for the data
            if(temp.data == data):
                break
            prev = temp
            temp = temp.next
            
        # if the data is not found
        if(temp == None):
            print(data, "does not exist in the LinkedList")
            return
            
        # exchange the links of prev.next and node.next
        prev.next = temp.next
        temp = None
            
            
    def display(self):
        temp = self.head
        print("The LinkedList is: ")
        while(temp):
            print(temp.data, end = " ")
            temp = temp.next
        print()
        
    

# DRIVER CODE

linked_list = LinkedList()
linked_list.addEnd(10)
linked_list.addEnd(20)
linked_list.addEnd(30)
linked_list.addEnd(40)
linked_list.addEnd(50)
linked_list.addEnd(60)
linked_list.display()

linked_list.addBegining(5)
linked_list.display()

linked_list.addMid(linked_list.head.next.next.next, 35)
linked_list.display()

linked_list.remove(5)
linked_list.display()
```
