```python
import sys
class MinHeap:
    # constructor
    def __init__(self, maxsize):
        self.maxsize = maxsize
        self.size = 0
        self.Heap = [0] * (self.maxsize + 1)
        # self.Heap[0] = sys.maxsize
        self.Heap[0] = (-1 * sys.maxsize)
        self.FRONT = 1
        
    def parent(self, index):
        return index//2
    
    def leftChild(self, index):
        return index*2
    
    def rightChild(self, index):
        return (2*index)+1
    
    def swap(self, index1, index2):
        self.Heap[index1], self.Heap[index2] = self.Heap[index2], self.Heap[index1]
    
    def insert(self, data):
        if(self.size >= self.maxsize):
            print("HEAP IS FULL!")
            return
        
        self.size += 1
        self.Heap[self.size] = data
        
        # check the properities
        temp = self.size
        while self.Heap[temp] < self.Heap[self.parent(temp)]:
            self.swap(temp, self.parent(temp))
            temp = self.parent(temp)
            
    def minHeapify(self, index):
        if not (index >= ((self.size//2)+1)):
            if (self.Heap[index] > self.Heap[self.leftChild(index)] or self.Heap[index] > self.Heap[self.rightChild(index)]):
                if self.Heap[self.leftChild(index)] < self.Heap[self.rightChild(index)]:
                    self.swap(index, self.leftChild(index))
                    self.minHeapify(self.leftChild(index))
                else:
                    self.swap(index, self.rightChild(index))
                    self.minHeapify(self.rightChild(index))
            
            
    def remove(self):
        deleted = self.Heap[self.FRONT]
        self.Heap[self.FRONT] = self.Heap[self.size]
        self.size -= 1
        self.minHeapify(self.FRONT)
        return deleted
    
    def display(self):
        for i in range(1, (self.size//2)+1):
            print("Parent:", str(self.Heap[i]),
                 "Left-Child:", str(self.Heap[self.leftChild(i)]),
                 "Right-Child:", str(self.Heap[self.rightChild(i)]))
            

# DRIVER CODE
print("Heap is as follows:")
minHeap = MinHeap(10)
minHeap.insert(23)
minHeap.insert(45)
minHeap.insert(12)
minHeap.insert(78)
minHeap.insert(45)
minHeap.insert(43)
minHeap.insert(4)
minHeap.insert(20)

minHeap.display()

print("Removed the data: ", minHeap.remove(), "\n")

print("After deletion, heap is as follows: \n")
minHeap.display()
```
