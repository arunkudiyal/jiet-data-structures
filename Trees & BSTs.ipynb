```python
# Binary Tree

class Node:
    def __init__(self, key):
        self. key = key
        self.left = None
        self.right = None
      
# LDR
def inOrder(node):
    if(not node):
        return
    inOrder(node.left)
    print(node.key, end = " ")
    inOrder(node.right)

# DLR
def preOrder(node):
    if(node):
        print(node.key, end = " ")
        preOrder(node.left)
        preOrder(node.right)

# LRD
def postOrder(node):
    if(node):
        postOrder(node.left)
        postOrder(node.right)
        print(node.key, end = " ")
        
# Geneate a Binary tree
root = Node("A")
root.left = Node("B")
root.right = Node("C")
root.left.left = Node("D")
root.left.right = Node("E")
root.right.left = Node("F")
root.right.right = Node("G")

print("In-Order is")
inOrder(root)
print()

print("Pre-Order is")
preOrder(root)
print()

print("Post-Order is")
postOrder(root)
print()
```

```python
# Binary Search Tree


class Node:
    def __init__(self, key):
        self. key = key
        self.left = None
        self.right = None
      
# LDR
def inOrder(node):
    if(node):
        inOrder(node.left)
        print(node.key, end = " ")
        inOrder(node.right)

# DLR
def preOrder(node):
    if(node):
        print(node.key, end = " ")
        preOrder(node.left)
        preOrder(node.right)

# LRD
def postOrder(node):
    if(node):
        postOrder(node.left)
        postOrder(node.right)
        print(node.key, end = " ")
        

# insert(root, key)
def insert(root, key):
    if root is None:
        return Node(key)
    
    else:
        # key < root.data | key > root.data
        if key < root.key:
            root.left = insert(root.left, key)
        elif key > root.key:
            root.right = insert(root.right, key)
        else:
            return root
    return root

def minValue(node):
    current = node
    while(current.left is not None):
        current = current.left
    return current

# remove(root, key):
def remove(root, key):
    # Base case
    if root is None:
        return
    
    # search the key
    if key < root.key:
        root.left = remove(root.left, key)
    elif key > root.key:
        root.right = remove(root.right, key)
        
    else:
        
        # one child
        if root.left is None:
            temp = root.right
            root = None
            return temp
        
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        
        # two children
        temp = minValue(root.right)
        root.key = temp.key
        root.right = remove(root.right, temp.key)
    return root

# DRIVER METHOD

root = Node(10)
root = insert(root, 22)
root = insert(root, 7)
root = insert(root, 20)
root = insert(root, 70)
root = insert(root, 30)

inOrder(root)
print()

root = remove(root, 10)
inOrder(root)
print()

root = remove(root, 7)
inOrder(root)
print()
```
